### Question 1

```python
#Importing the dataset

movies = pd.read_csv(r"C:\Users\admin\Downloads\Assignment\MoviesData.csv",index_col=0)
movies
```

### Question 2

```python
#Generating descriptive statistics
# we can do it using .describe()

movies.describe()
```


### Question 3

```python
#Finding top-rated movies produced in the United States that have a PG-13 rating

movies[(movies['content_rating'] == 'PG-13')  & (movies['Country'] == 'USA')]
```

### Question 4

```python
#Finding out whether any of the top-rated movies produced in 2014 were not produced in the United States

movies[(movies['title_year'] == 2014) & (movies['Country'] != 'USA')]
```

### Question 5

```python
# Find the percentage of the top-rated movies that are in genres 3

percentage_genre3 = (movies['genre_3'].count() / len(movies)*100)
message3 = "The percentage of the top-rated movies that are in genres 3 is : {0} %.".format(percentage_genre3)
print(message3)
```

```python
# Find the percentage of the top-rated movies that are in  genres 2

percentage_genre2 = ((movies['genre_2'].count() - movies['genre_3'].count())/ len(movies)) * 100
message2 = "The percentage of the top-rated movies that are in genres 2 is : {0} %.".format(percentage_genre2)
print(message2)
```

```python
# Find the percentage of the top-rated movies that are in  genres 1

percentage_genre1 = ((movies['genre_1'].count() - movies['genre_2'].count() )/ len(movies)) * 100
message1 = "The percentage of the top-rated movies that are in genres 1 is : {0} %.".format(percentage_genre1)
print(message1)
```

### Question 6

```python
# Convert the budget values from “dollars” to “dollars in millions” for all top-rated movies. Round the converted values down to 3 decimal places. 

round(movies['budget'] / 1000000,3) 
```

```python
# Convert the gross values from “dollars” to “dollars in millions” for all top-rated movies. Round the converted values down to 3 decimal places. 

round(movies['Gross'] / 1000000,3) 
```

### Question 7

```python
# Listing all details for the top 10 movies with the highest profit, sorted from highest to lowest.

movies['Profit'] =  movies['Gross'] - movies['budget']

movies_top_10 = movies.sort_values(by='Profit',ascending = False)
movies_top_10.head(10)
```

### Question 8

```python
# Generate a list of all the actors, in alphabetical order by the first name, that have starred in a top-rated movie.

actors1 = movies['actor_1_name']
actors2 = movies['actor_2_name']
actors3 = movies['actor_3_name']

all_actors = pd.concat( [actors1, actors2, actors3]).sort_values()
all_actors_list = list(all_actors.unique())
all_actors_list.remove('0')

all_actors_list
```

### Question 9

```python
# the top 3 actors who appeared in the most top-rated movies. (Based on rating IMDb_rating column.)

top3_rated_movies  = movies.nlargest(3, ['IMDb_rating'])
top3_rated_movies['actor_1_name']
```

### Question 10

```python
# data visualization that shows each country and the number of top-rated movies produced in it

movies_viz = movies.groupby(['Country']).count()[['title_year']].sort_values(by='title_year',ascending = False)
movies_viz.plot(kind = 'bar')

plt.title(' Country and Number of Top-rated Movies produced')
plt.xlabel('Countries')
plt.ylabel('Number of Movies produced')
```
